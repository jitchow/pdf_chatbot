```python
import gradio as gr
import os
os.environ["no_proxy"] = "localhost,127.0.0.1,::1"

def greet(name):
    return "Hello " + name + "!"

demo = gr.Interface(fn=greet, inputs="text", outputs="text")
    
if __name__ == "__main__":
    demo.launch(show_api=False)   
```
